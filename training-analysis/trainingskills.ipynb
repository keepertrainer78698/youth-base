```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/matthiashugli/Downloads/s1_core_trm_player_unit_skills.csv', sep=';', header=[1,2], skiprows=[3])
```

```python
df.head()
```

```python
melt_df = df.iloc[:, :17] \
            .set_index([('Unnamed: 0_level_0', 'Unnamed: 0_level_1'), ('Unnamed: 1_level_0', 'Unnamed: 1_level_1'), ('Parent Category Id', 'Category Name')])   \
                .melt(var_name=['Kategorie', 'Skill'], value_name='Dauer', ignore_index=False) \
                    .dropna(subset=['Dauer']) \
                        .reset_index()

melt_df.columns = ['Keeper', 'Team', 'DatumZeit', 'Kategorie', 'Skill', 'Dauer']
```

```python
pivot_df = melt_df.pivot_table(index=['Keeper', 'Team', 'DatumZeit', 'Dauer'], columns='Kategorie', values=['Skill'], aggfunc=', '.join).reset_index()

pivot_df[('Skill', 'DEF')] = pivot_df[('Skill', 'DEF')].str.split(',')
pivot_df[('Skill', 'OFF')] = pivot_df[('Skill', 'OFF')].str.split(',')
pivot_df[('Skill', 'Verhalten')] = pivot_df[('Skill', 'Verhalten')].str.split(',')

pivot_df = pivot_df.explode(('Skill', 'DEF'))
pivot_df = pivot_df.explode(('Skill', 'OFF'))
pivot_df = pivot_df.explode(('Skill', 'Verhalten'))
# pivot_df.reset_index(inplace=True)

pivot_df.columns = ['Keeper', 'Team', 'DatumZeit', 'Dauer', 'DEF', 'OFF', ' Verhalten']

pivot_df.to_csv('/Users/matthiashugli/Downloads/gk_skill.csv', index=False)
```
